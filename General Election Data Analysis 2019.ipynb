```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (15,10)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
path = '/content/Election data.csv'
```

```python
election_data = pd.read_csv(path)
```

```python
election_data
```

```python
election_data.at[172, "WINNER"] = 0
election_data.at[192, "WINNER"] = 1
election_data.at[702, "WINNER"] = 1
election_data.at[951, "WINNER"] = 1
election_data.at[1132, "WINNER"] = 1
```

```python
election_data.at[election_data.EDUCATION == 'Post Graduate\n','EDUCATION'] = 'Post Graduate'
election_data.at[election_data.EDUCATION == 'Graduate Professional', 'EDUCATION'] = 'Graduate\nProfessional'
```

```python
election_data.columns
```

```python
election_data.rename(columns = {'CRIMINAL\nCASES' : 'CRIMINAL CASES', 'TOTAL\nVOTES' : 'TOTAL VOTES'}, inplace = True)
election_data.sort_values(['STATE', 'CONSTITUENCY'], inplace = True)
```

```python
election_data.columns
```

```python
election_data
```

```python
election_data.info()
```

```python
election_data["CRIMINAL CASES"] = pd.to_numeric(election_data["CRIMINAL CASES"], errors = 'coerce').convert_dtypes()
```

```python
candidates_data = election_data[election_data.NAME != 'NOTA']
```

```python
candidates_data
```

```python
candidates_data.columns
```

```python
candidates_data = candidates_data.drop(['SYMBOL', 'GENERAL\nVOTES', 'POSTAL\nVOTES', 'TOTAL VOTES','OVER TOTAL ELECTORS \nIN CONSTITUENCY',
                                        'OVER TOTAL VOTES POLLED \nIN CONSTITUENCY', 'TOTAL ELECTORS'], axis = 1)
```

```python
candidates_data
```

```python
candidates_data.describe()
```

```python
winners = candidates_data[candidates_data.WINNER == 1].sort_values(['STATE', 'CONSTITUENCY']).reset_index()
```

```python
winners.drop(['index', 'WINNER'], axis = 1, inplace = True)
```

```python
winners
```

```python
candidates_data.PARTY.unique()
```

```python
candidates_data[candidates_data.PARTY == 'NOTA'].NAME
```

```python
print('No. of parties fielded atleast one candidate = ', election_data.PARTY.unique().shape[0]-2)
                                                                                        # -1 for IND 
                                                                                        # -1 for NOTA 
```

```python
print('No. of Independent Candidates Contested = ', candidates_data[candidates_data.PARTY == 'IND'].shape[0])
```

```python
print('No. of parties won atleast one seat in contested = ', winners.PARTY.unique().shape[0]-1)
```

```python
print('No. of independent candidates won the election = ', winners[winners.PARTY == 'IND'].shape[0])
```

```python
party_wise_seats = winners.PARTY.value_counts().sort_values(ascending = False)
```

```python
party_wise_seats
```

```python
others = party_wise_seats[party_wise_seats<10].sum()
```

```python
others
```

```python
seat_distribution = party_wise_seats[party_wise_seats>=10].append(pd.Series({'Others' : others}))
```

```python
pd.DataFrame(seat_distribution)
```

```python
plt.figure(figsize = (15,10))
plt.title('Party wise Seats gained')

plt.pie(seat_distribution, labels = seat_distribution.index, 
        colors = ['#f45503', '#1df903', '#031273', '#515db4', '#a1ff82', '#fcd882', '#fc00f4', '#31f8f4', '#01f858', '#a06601'],
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1.0});

seat_percent = round((seat_distribution/seat_distribution.sum())*100, 2)
legend = seat_percent.index + " (" + seat_percent.values.astype(str) + " % ) " + seat_distribution.values.astype(str)
plt.legend(legend, loc = "center", bbox_to_anchor = (1.5,0.5));
```

```python
plt.figure(figsize = (20,10))
plt.title('Age of Candidates Contested and Won', fontsize = 20)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('No. of Candidates', fontsize = 15)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

sns.histplot(data = candidates_data, x='AGE', bins = np.arange(20,100,5), color = 'blue', alpha = 0.5)
sns.histplot(data = winners, x = 'AGE', bins = np.arange(20,100,5), color = 'green', alpha = 1.0)

plt.legend(['Candidates Contested', 'Candidates Won'], fontsize = 20)

plt.text(80,240, 'All Candidates: ')
plt.figtext(0.73, 0.63, round(candidates_data.describe().AGE[['mean', 'min', 'max']], 2).to_string()) 

plt.text(80, 185, 'WinningCandidates: ')
plt.figtext(0.73, 0.5, round(winners.describe().AGE[['mean', 'min', 'max']], 2).to_string());
```

```python
print('Youngest Member of House: ')
winners[(winners.AGE == 25)][["NAME", "PARTY", "STATE", "CONSTITUENCY", "GENDER"]].reset_index(drop = True)
```

```python
print('Oldest Member of House: ')
winners[(winners.AGE == 86)][["NAME", "PARTY", "STATE", "CONSTITUENCY", "GENDER"]].reset_index(drop = True)
```

```python
gender_group = candidates_data.groupby(['GENDER', 'WINNER']).size()
gender_group = gender_group.unstack()
gender_group = gender_group[[1,0]]

sns.set_palette(sns.color_palette('icefire'))

gender_group.plot(kind = 'barh', figsize = (15,6), title = 'Gender comparision of Contested and winning candidates')

plt.legend(['Won', 'Lost'])
plt.xlabel('No. of Candidates')
plt.ylabel('Gender')

plt.figtext(0.738,0.53,'Contesting Candidates\n'+
            round((candidates_data.GENDER.value_counts(normalize=True)*100),2).to_string().replace("\n", "%\n")+"%")

plt.figtext(0.738,0.33,'Winning Candidates\n'+
            round((candidates_data.GENDER.value_counts(normalize=True)*100),2).to_string().replace("\n", "%\n")+"%")

win_percent = round((winners.GENDER.value_counts()/candidates_data.GENDER.value_counts())*100,2)

plt.figtext(0.41, 0.63, str(round(win_percent.MALE, 2))+ " % Candidates Won")
plt.figtext(0.18, 0.25, str(round(win_percent.FEMALE, 2))+ " % Candidates Won")
```
